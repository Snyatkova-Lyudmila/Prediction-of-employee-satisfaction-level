## Введение

**Описание проекта**

HR-аналитики компании «Работа с заботой» помогают бизнесу оптимизировать управление персоналом: бизнес предоставляет данные, а аналитики предлагают, как избежать финансовых потерь и оттока сотрудников. В этом HR-аналитикам пригодится машинное обучение, с помощью которого получится быстрее и точнее отвечать на вопросы бизнеса.

Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён. 
Собирать данные такими опросниками не так легко: компания большая, и всех сотрудников надо сначала оповестить об опросе, а затем проследить, что все его прошли. 

**Первая задача** — построить модель, которая сможет предсказать уровень удовлетворённости сотрудника на основе данных заказчика. 
Почему бизнесу это важно: удовлетворённость работой напрямую влияет на отток сотрудников. А предсказание оттока — одна из важнейших задач HR-аналитиков. Внезапные увольнения несут в себе риски для компании, особенно если уходит важный сотрудник.

**Вторая задача** — построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.

## Задача 1 / Шаг 3. Исследовательский анализ данных

>В целях оптимизации кода для построения графиков количественных признаков (графика распределения и диаграммы размаха) создадим функцию **graphs_col**:  

* на входе функция принимает название датасета (df), название признака (feature) и количество корзин для построения графика распределения (number_bins)  
* результат работы функции - график распределения и диаграмма размаха для указанного признака

```python
def graphs_col (df, feature, number_bins, name):
    
    plt.figure(figsize=(20, 6))
    plt.suptitle(f'Графики для датафрейма {name}', fontsize = 15, fontweight='bold')
        
    plt.subplot(1, 2, 1)
    plt.hist(df[feature], edgecolor='black', bins=number_bins)
    plt.title(f'Распределение признака {feature}', fontsize=15)
    plt.ylabel('плотность распределения', fontsize=14)
    plt.xlabel(f'{feature}', fontsize=14)

    plt.subplot(1, 2, 2)
    df.boxplot(column=feature, fontsize=14)
    plt.title(f'Диаграмма размаха для параметра {feature}', fontsize=15)
    plt.ylabel(f'{feature}', fontsize=14)
    plt.show()
```

>В целях оптимизации кода для построения графиков категориальных признаков (столбчатой и круговой диаграмм) создадим функцию **graphs_cat**:

* на входе функция принимает название датасета (df) и название признака (feature)  
* результат работы функции - столбчатая и круговая диаграмма для указанного признака

```python
def graphs_cat (df, feature, name):
    
    palette = sns.color_palette("tab10", len(df))
    plt.figure(figsize=(20, 6))
    plt.suptitle(f'Графики для датафрейма {name}', fontsize = 15, fontweight='bold')

    plt.subplot(1, 2, 1)
    df.groupby(feature)[feature].count().sort_values(ascending=False).plot(kind='bar')
    plt.title(f'Столбчатая диаграмма: {feature}', fontsize=15)
    plt.xticks(rotation=0)
    plt.xlabel(f'{feature}', fontsize=14)
    plt.ylabel('количество', fontsize=14)

    plt.subplot(1, 2, 2)
    plt.title(f'Круговая диаграмма: {feature}', fontsize=15)
    df.groupby(feature)[feature].count().sort_values(ascending=False).plot(
                                    kind='pie',
                                    colors=palette, autopct='%1.1f%%',
                                    subplots=True,
                                    ylabel=' ')
    plt.show()
```

## Задача 1 / Шаг 4. Подготовка данных и обучение моделей

>Для оценки качества моделей напишем функцию **smape**, которая будет вычислять метрику SMAPE (симметричное среднее абсолютное процентное отклонение):

```python
def smape(y_true, y_pred):
    return 100 * np.mean(np.abs(y_pred - y_true) / ((np.abs(y_pred) + np.abs(y_true))/2))
```

Перед тем, как разделить тренировочную выборку на две (с целевым признаком и без него) удалим из данных неинформативный признак 'id'.

Вспомним, что при обработки данных были обнаружены пропущенные значения в столбцах 'dept' (отдел, в котором работает сотрудник) и 'level' (уровень занимаемой должности). Подготовку признаков выполним в пайплайне, дополнив пайплайн шага предобработки - на этом этапе избавимся от пропусков.

Признаки - категориальные, поэтому логичным решением будет заменить отсутствующие значения наиболее часто встречающимся значением в столбце (strategy='most_frequent').

При кодировании будем учитывать особенности признаков и моделей. Будем использовать два кодировщика: OneHotEncoder, OrdinalEncoder, проводить обучение двух моделей: линейной и дерева решений.

```python
# создание констант
RANDOM_STATE = 42
```

```python
# создадим списки с названиями признаков
# признаки для OneHotEncoder
ohe_columns = ['last_year_violations', 'last_year_promo', 'dept']

# признаки для OrdinalEncoder
ord_columns = ['level', 'workload']

# признаки для масштабирования
num_columns = ['salary', 'employment_years', 'supervisor_evaluation'] 
```

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE

ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore')
        )
    ]
)
```

```python
# cоздаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE

ord_pipe = Pipeline(
    [
        (
            'simpleimputer_before_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ), 
        (
            'ord',
            OrdinalEncoder(categories=[
                ['junior', 'middle','senior'],
                ['low', 'medium', 'high']
            ], handle_unknown='use_encoded_value', unknown_value=np.nan)
        ), 
        (
            'simpleimputer_after_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

```python
# создаём общий пайплайн для подготовки данных

data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель

pipeline_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=RANDOM_STATE))
])
```

```python
 # список словарей

param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 15),
        'models__min_samples_split': range(2, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # словарь для модели LinearRegression()
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```

Создадим свою метрику SMAPE, используя функцию smape:

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
# SMAPE - ошибка, её минимизируем
```

Для экономии ресурсов и ускорения работы алгоритма для оптимизации гиперпараметров модели будем использовать метод RandomizedSearchCV:

```python
random_search = RandomizedSearchCV(
    pipeline_final, 
    param_grid, 
    cv=5,
    scoring=smape_scorer,
    random_state=RANDOM_STATE,
    n_jobs=-1
)
```

```python
random_search.fit(X_train, y_train)
```

```python
print('Лучшая модель и её параметры:\n\n', random_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', np.abs(random_search.best_score_))
```

```python
# проверим работу модели на тестовой выборке
# рассчитаем прогноз на тестовых данных
y_pred = random_search.best_estimator_.predict(X_test_)
print(f'Значение метрики SMAPE на тестовых данных: {smape(y_test_, y_pred)}')
```

На этом этапе составим портрет «уволившегося сотрудника». Также сравним среднее значение зарплаты ушедших сотрудников с теми, кто остался в компании. Для анализа построим графики, на которых результат будет представлен наглядно в сравнении по всем параметрам. Будем рассматривать cоотношение уволившихся и работающих сотрудников и cоотношение зарплаты уволившихся и работающих.

>Для удобства при анализе соотношения уволившихся и работающих сотрудников создадим функцию **salary_level**:  

* на входе функция принимает сумму зарплаты сотрудника (value)  
* результат работы функции - уровень зарплаты (выделим пять уровней: от 80 тыс, 60-80 тыс, 40-60 тыс, 20-40 тыс и до 20 тыс.)

```python
def salary_level(value):
    if value >= 80000:
            return "от 80к"
    elif value < 80000 and value >= 60000:
            return "60-80к"
    elif value < 60000 and value >= 40000:
            return "40-60к"
    elif value < 40000 and value >= 20000:
            return "20-40к"
    elif value < 20000:
            return "до 20к"    
 
```

Для анализа в датафрейм train1 добавим новый столбец 'salary_level', в котором будет указан уровень зарплаты сотрудника.  
Чтобы не портить датафрейм train1, создадим новый датафрейм train2 (он потребуется нам только для графика "Соотношение уволившихся и работающих сотрудников"). 

```python
# создадим копию датасета
train2 = train1.copy(deep=True)
train2['salary_level'] = train2['salary'].apply(salary_level)
train2.head()
```

## Задача 2 / Шаг 5. Подготовка данных и обучение моделей

На этом этапе проведем обучение четырех моделей, к которым подберем гиперпараметры.  
Качество лучшей модели проверим с помощью метрики ROC-AUC (критерий успеха: ROC-AUC >= 0.91 на тестовой выборке).

Подготовку данных будем проводить по анологии с первой задачей, только к масштабированию количественных признаков добавится новый признак ('job_satisfaction_rate_pred').

Также отметим, что целевой признак 'quit' - категориальный, для его кодирования будем использовать метод LabelEncoder.

---
Перед тем, как разделить тренировочную выборку на две (с целевым признаком и без него) удалим из данных неинформативный признак 'id'.  
Затем проверим данные на наличие дубликатов.

```python
train1_temp = train1.drop(['id'], axis=1)
print('Количество дубликатов:',train1_temp.duplicated().sum())
```

Оптимальным решением будет удалить дубликаты, так как такие данные не несут дополнительной информации в модель и придают необоснованный вес таким наблюдениям. Также увеличение объёма выборки увеличивает расход вычислительных ресурсов и времени обучения.

```python
# удалим дубликаты
train1_temp = train1_temp.drop_duplicates().reset_index(drop=True)
```

```python
# сохраним входные и целевой признаки
X_train1 = train1_temp.drop(['quit'], axis=1)
y_train1 = train1_temp['quit']
```

```python
display(X_train1.shape)
display(y_train1.shape)
```

---
Датафреймы X_test1 и y_test1 тестовой выборки соединим по 'id', чтобы данные целевого признака шли в том же порядке, что и данные входных.

```python
test1 = X_test1.merge(y_test1, on='id')
```

```python
X_test1_ = test1.drop(['quit','id'], axis=1)
y_test1_ = test1['quit']
```

```python
display(X_test1_.shape)
display(y_test1_.shape)
```

```python
# закодируем целевой признак
label_encoder = LabelEncoder()
y_train1 = label_encoder.fit_transform(y_train1)
y_test1_ = label_encoder.transform(y_test1_)
```

```python
# создадим списки с названиями признаков

# признаки для OneHotEncoder
ohe_columns = ['last_year_violations', 'last_year_promo', 'dept']

# признаки для OrdinalEncoder
ord_columns = ['level', 'workload']

# признаки для масштабирования
num_columns = ['salary', 'employment_years', 'supervisor_evaluation', 'job_satisfaction_rate_pred'] 
```

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE

ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore')
        )
    ]
)
```

```python
# cоздаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE

ord_pipe = Pipeline(
    [
        (
            'simpleimputer_before_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ), 
        (
            'ord',
            OrdinalEncoder(categories=[
                ['junior', 'middle','senior'],
                ['low', 'medium', 'high']
            ], handle_unknown='use_encoded_value', unknown_value=np.nan)
        ), 
        (
            'simpleimputer_after_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

```python
# создаём общий пайплайн для подготовки данных

data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель

pipeline_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

Итак, мы выполнили подготовку данных: кодирование, масштабирование, обработку пропусков с использованием пайплайна.  
Следующий шаг - обучение моделей.

Цеелвой признак - 'quit' (увольнение сотрудника из компании). Есть всего два варианта: уволят / не уволят.  
Таким образом, необходимо решить задачу бинарной классификации. 

Для бинарной классификации подойдут следующие модели:
* дерево решений
* метод k-ближайших соседей
* логистическая регрессия
* метод опорных векторов

```python
param_grid = [
    
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(4, 11),
        'models__max_features': range(6, 12),
        'preprocessor__num': [StandardScaler(), MinMaxScaler()],
    },
 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler()],
    },

    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 3),
        'preprocessor__num': [StandardScaler(), MinMaxScaler()],
    },

    {
        'models': [SVC(
            random_state=RANDOM_STATE,
            kernel = 'rbf'
        )],
        'preprocessor__num': [StandardScaler(), MinMaxScaler()],
    }
]
```

```python
randomized_search = RandomizedSearchCV(
    pipeline_final, 
    param_grid, 
    cv=5,
    n_iter=20,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train1, y_train1)
```

```python
print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)
```

```python
# проверим работу модели на тестовой выборке
# рассчитаем прогноз на тестовых данных
y_test_proba = randomized_search.predict_proba(X_test1_)[:,1]
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test1_, y_test_proba)}')
```

Значение 7означает, что при каждом разделении узла будет рассматриваться не более 11 признаков. Это может помочь уменьшить сложность модели и сделать ее менее прогрессивным переобучением.

```python
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE

ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore')
        )
    ]
)
```

```python
# cоздаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE

ord_pipe = Pipeline(
    [
        (
            'simpleimputer_before_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ), 
        (
            'ord',
            OrdinalEncoder(categories=[
                ['junior', 'middle','senior'],
                ['low', 'medium', 'high']
            ], handle_unknown='use_encoded_value', unknown_value=np.nan)
        ), 
        (
            'simpleimputer_after_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

```python
# создаём общий пайплайн для подготовки данных

data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

```python
# создаём итоговый пайплайн: подготовка данных и модель

pipeline_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
randomized_search = RandomizedSearchCV(
    pipeline_final, 
    param_grid, 
    cv=5,
    n_iter=20,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train2, y_train1)
```

```python
print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)
```

```python
# проверим работу модели на тестовой выборке
# рассчитаем прогноз на тестовых данных
y_test_proba = randomized_search.predict_proba(X_test2)[:,1]
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test1_, y_test_proba)}')
```

**РЕКОМЕНДАЦИИ ДЛЯ БИЗНЕСА**

Внезапные увольнения - это всегда большие риски для компании, особенно если уходит важный сотрудник.  
Для того, чтобы снизить количество увольнений в компании, можно предложить заказчику следующие рекомендации:

1) **отслеживать (и стремиться повышать) удовлетворенность работой сотрудников**

* *проведение опросов удовлетворенности персонала*: регулярные (1-2 раза в год) анономные опросы, где сотрудники могут свободно говорить правду без страха последствий, открыто высказывать свое мнение

* *создание здорового микроклимата*: обеспечение комфортного и поддерживающего рабочего места, где каждый сотрудник чувствует себя услышанным и уважаемым; создание открытой среды для обсуждения проблем и предложений, развитие корпоративной культуры

* *анализ содержания труда*: удовольствие от выполнения задач напрямую связано с качеством работы и продуктивностью - создание интересного и значимого наполнения работы помогает повысить уровень удовлетворенности сотрудников

* *создание комфортных условий труда*: рабочее место, техническое оснащение, освещенность и уровень шума

* *применение персонализированного подхода*: персонализация взаимодействия руководства с сотрудниками, проявление внимания к их потребностям и интересам, а также использование интерактивного формата общения

2) **пересмотреть условия для повышения уровня должности для наиболее перспективных сотрудников**  

* *быстрое продвижение для наиболее перспективных сотрудников*: особенно в начале карьеры наиболее перспективных сотрудников быстрее переводить из разряда junior в разряд medium

* *мотивирующие вознаграждения*: дополнительные вознаграждения / бонусы для сотрудников уровня junior, перешедших на уровень medium за короткие сроки

* *определение критериев продвижения*: четкое определение критериев, по которым сотрудник может быть повышен (это могут быть технические навыки, опыт работы, успехи в проектах, уровень ответственности и другие аспекты); критерии должны быть объективными и прозрачными для всех сотрудников

* *разработка программы развития карьеры*: создание программы развития карьеры, которая будет включать в себя планы обучения, менторство, участие в проектах высокого уровня сложности и другие возможности для профессионального роста, что поможет сотрудникам видеть путь к продвижению и понять, какие действия они должны предпринять для достижения своих целей

* *частая обратная связь*: регулярная и конструктивная обратная связь с руководством и коллегами поможет сотрудникам понимать, в чем они хорошо справляются, и в чем есть места для улучшения, что также даст возможность для корректировки планов развития и обучения

* *поддержка руководства*: руководство должно активно поддерживать процессы продвижения и развитие сотрудников, участвуя в их обучении, предоставлении обратной связи и демонстрируя личный пример в отношении профессионального роста и самосовершенствования 

**3) увеличить загруженность работой сотрудника**

* *эффективное управление загруженностью*: правильное распределение задач в команде и контроль над их выполнением

* *планирование и прогнозирование*: знание текущей загрузки сотрудников и прогнозирование будущих задач поможет оптимизировать распределение ресурсов

* *использование инструментов управления работой*: инструменты (такие как, например, Asana) позволяют всем членам команды видеть статусы задач и их место в общем процессе

* *правильное распределение и постановка задач*: четкая формулировка задания, использование методологии SMART (конкретные, измеримые, достижимые, значимые и ограниченные по времени); проверка понимания сотрудниками новой задачи, контроль ее выполнение, получение обратной связи

* *мотивация*: замотивировать сотрудника - побудить его к действию и вызвать желание работать (дополнительные вознаграждения / бонусы или участие в новых интересных проектах)  

**4) Пересмотреть размер вознаграждения для наиболее перспективных сотрудников с небольшим стажемм**  
У сотрудника есть рост в должности и он загружен работой - соответственно, растет и зарплата. Но рекомендуем рассмотреть возможность *выплат премий / повышения уровня зарплаты наиболее ценным сотрудникам, чей стаж в компании меньше года*.

**5) Особое внимание рекомендуем уделять молодым сотрудниками - тем, кто работает в компании меньше года, а также сотрудникам отдела продаж**.
